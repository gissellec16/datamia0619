```python
# your code here
uf1 = urllib.request.urlopen('https://www.zalando.fr/api/catalog/articles?categories=promo-femme&limit=84&offset=84&sort=sale')
response1 = uf1.read().decode('utf-8')
results1 = json.loads(response1)


flattened_data1 = json_normalize(results1)

flattened_data11 = json_normalize(flattened_data1.articles[0])
flattened_data11

uf2 = urllib.request.urlopen('https://www.zalando.fr/api/catalog/articles?categories=promo-homme&limit=84&offset=84&sort=sale')
response2 = uf2.read().decode('utf-8')
results2 = json.loads(response2)


flattened_data2 = json_normalize(results2)

flattened_data22 = json_normalize(flattened_data2.articles[0])
flattened_data22
```
