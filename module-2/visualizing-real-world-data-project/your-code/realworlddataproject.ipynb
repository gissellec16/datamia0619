```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.plotly as py
import cufflinks as cf
```

```python
data=pd.read_csv('/Users/gissellec/Downloads/abalone.csv')
```

```python
data.columns
```

```python
cf.go_offline()
```

```python
data.iplot(x='diameter', y='height', categories='sex',
           xTitle='Diameter', yTitle='Height',
           title='Diameter vs Height of Abalones')
```

The scatterplot is meant to show the relationship between the height of the abalone and its diameter, with color coding meant to show the difference between the males and females of the species. 

```python
df = data['whole weight']
df.iplot(kind='hist', xTitle='whole weight', yTitle='Count', 
           title='Whole Weight Distribution')
```

This histogram is meant to show the frequencies of the whole weights recorded in this dataset. 

```python
data.boxplot(column="shell weight")
```

This boxplot shows the distribution of the shell weights of all the alabones, and we can see that there are many upper limit outliers. 

```python
ax = sns.boxplot(x="sex", y="shell weight", data=data)
```

These boxplots show shell weights as well, only broken up by sex. We can see the upper limit outlier issue is present across the recorded sexes. 

```python
data_bar=data[["whole weight", "shucked weight", "viscera weight","shell weight", "sex"]]
data_mean=data_bar.groupby("sex").agg("mean")
data_mean.plot.bar(figsize=(20,10))
```

The bar graph shows the various recorded weights from the dataset (whole, shucked, viscera, shell) and breaks them up by sex. This breakout allows the viewer to see how the weights relate to each other by sex as well as across sexes.
